## Passo 1

Importar libs e carregar matplotlib inline para o jupyter.

```python
%matplotlib inline
import pandas as pd
import matplotlib as plt
```

## Passo 2

Carregar o csv com as informações sobre os campeonatos de league of legends e visualizar os dados.

```python
data = pd.read_csv('csv/summer-2019-lol-leagues.csv')

data
```

O dataframe possui 12612 linhas e 98 colunas. Possui informações de campeonatos da europa, coreia, china, estados unidos, etc. Portanto é preciso filtrar apenas os dados relacionados ao campeonato brasileiro.

## Passo 3

filtrar e limpar os dados.

```python
cblol_data = data[[
    'gameid',
    'result',
    'league', 
    'split', 
    'week', 
    'game', 
    'playerid', 
    'side', 
    'position', 
    'player', 
    'team',
    'champion',
    'ban1',
    'ban2',
    'ban3',
    'ban4',
    'ban5',
    'gamelength',
    'teamkills',
    'teamdeaths',
    'fbtime',
    'teamtowerkills',
    'wards',
    'visionwards',
    'totalgold',
]]

cblol_data = cblol_data[cblol_data['league'] == 'CBLoL']

cblol_data
```

A tabela acima são todos os jogos que aconteceram durante a temporada. Observe que no dataset, os jogos são representados por playerid 100 ou 200.

## Passo 4

Agora vamos mostrar algumas informações relacionado a temporada.

```python
winners = team_data[team_data['result'] == 1]

winners['side'].value_counts().plot(kind='bar')
```

O jogo possui dois lados: azul e vermelho. A maioria das vitórias foram do lado azul nessa temporada.

```python
winner_data = team_data[team_data['result'] == 1]

print('Média da duração das partidas:', winner_data['gamelength'].mean())
print('Média de visão normal colocada no mapa:', team_data['wards'].mean())
print('Média de visão detectora colocada no mapa:', team_data['visionwards'].mean())
```

Em média, as partidas duraram 32.39 minutos, foram colocadas em média 112 wards comuns e 40 wards detectora no mapa.

```python
bans_data = team_data[['ban1', 'ban2', 'ban3', 'ban4', 'ban5']]

count_ban_data = bans_data.apply(pd.Series.value_counts).fillna(0)

count_ban_data
```

```python
count_ban_data['total'] = count_ban_data.sum(axis = 1)

sorted_count_ban_data = count_ban_data.sort_values(by='total', ascending=False)

sorted_count_ban_data.head(10)
```

A tabela mostra a quantidade que um campeão foi banido pela primeira, segunda, terceira, quarta e quinta rotação
e por último mostra o total de bans feito naquele campeão.

```python
sorted_count_ban_data['total'].head(10).plot(kind='bar')
```

O gráfico acima mostra os 10 campeões mais banidos da temporada.

```python
player_data = cblol_data[(cblol_data['playerid'] != 100) & (cblol_data['playerid'] != 200)]

winner_player_data = player_data[player_data['result'] == 1]
loser_player_data = player_data[player_data['result'] == 0]

champions_winner_season_count = winner_player_data['champion'].value_counts()
champions_loser_season_count = loser_player_data['champion'].value_counts()
```

```python
frame = { 'victory': champions_winner_season_count, 'defeat': champions_loser_season_count }

win_and_lose_df = pd.DataFrame(frame)
win_and_lose_df.fillna(0, inplace=True)

win_and_lose_df['total'] = win_and_lose_df.sum(axis=1)
win_and_lose_df['win_rate'] = (win_and_lose_df['victory'] / win_and_lose_df['defeat']) * 100

win_and_lose_df.sort_values(by='victory', ascending=False)
```

```python
win_and_lose_df[['victory', 'defeat']].head(10).plot(kind='bar')
```

O gráfico acima mostra os 10 campeões com mais vitória. Veja que assim como Aatrox foi um campeão com muita vitória,
também foi um campeão com um grande número de derrotas.

```python
win_and_lose_df[win_and_lose_df['total'] > 10].sort_values(by=['total', 'win_rate'], ascending=False)
```

Alguns campeões apresentam poucas partidas e 100% de win rate. Isso não me diz muito sobre o campeão já que ele teve duas ou três partidas.

As semifinais são jogos em melhor de cinco. Por essa informação podemos concluir que os dois finalistas foram Flamengo e INTZ.

Vamos verificar a quantidade de visão que os dois time colocaram no mapa durante toda a temporada.

```python
points_wards_info = championship_points_table[['wards', 'visionwards']]
sf_wards_info = sf_phase[['wards', 'visionwards']]
f_wards_info = f_phase[['wards', 'visionwards']]

joined_info = points_wards_info \
    .join(sf_wards_info, how='right', on='team', rsuffix='_semifinal') \
    .join(f_wards_info, how='right', on='team', rsuffix='_final')

joined_info
```

```python
joined_info.sum(axis=1)
```

Veja que a INTZ foi o time que mais pegou visão do mapa dos dois finalistas. Porém, o Flamengo ainda assim conseguiu o título.

```python
points_kill_info = championship_points_table[['teamkills']]
sf_kill_info = sf_phase[['teamkills']]
f_kill_info = f_phase[['teamkills']]

joined_kill_info = points_kill_info \
    .join(sf_kill_info, how='right', on='team', rsuffix='_semifinal') \
    .join(f_kill_info, how='right', on='team', rsuffix='_final')

joined_kill_info
```

```python
joined_kill_info.sum(axis=1)
```

Enquanto o Flamengo teve o maior número de abates.
